Exercise2.2 A sawtooth signal has a waveform that ramps up linearly from -1 to 1, then drops to -1 and repeats. See http://en.wikipedia.org/wiki/ Sawtooth_wave 

Write a class called SawtoothSignal that extends Signal and provides evaluate to evaluate a sawtooth signal. Compute the spectrum of a sawtooth wave. How does the harmonic structure compare to triangle and square waves?


연습 문제 2.2 톱니파 신호는 -1에서 1까지 선형으로 상승한 후 -1로 떨어지고 반복됩니다. http://en.wikipedia.org/wiki/ Sawtooth_wave를 참조하십시오.

Signal을 확장하는 SawtoothSignal이라는 클래스를 작성하고 톱니파 신호를 평가하기위한 평가를 제공합니다. 톱니파의 스펙트럼을 계산합니다. 고조파 구조는 삼각형과 사각파와 어떻게 비교됩니까?

```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot
import numpy as np
import math

import warnings
warnings.filterwarnings('ignore')

from IPython.html.widgets import interact, interact_manual, fixed
from IPython.html import widgets
from IPython.display import display

PI2 = 2 * math.pi
```

```python
class SawtoothSignal(thinkdsp.Sinusoid):

    
    def evaluate(self, ts):

        cycles = self.freq * ts + self.offset / PI2 #위상이 얼마나 되는지 계산
        frac, _ = np.modf(cycles) #정수 값 무시
        #ys = np.abs(frac - 0.5) -> 삼각파 만들때 필요한 식 : abs-절대값으로 만들어줌, frac-0.5-중간으로 옮겨줌
        #하지만 톱니파를 만들 때는 필요없다
        ys = thinkdsp.normalize(thinkdsp.unbias(frac), self.amp) #평준화 해준다.
        return ys
```

```python
sawtooth = SawtoothSignal().make_wave(duration=0.01, framerate=10000)
```

```python
sawtooth.plot()
```

```python
sawtooth = SawtoothSignal().make_wave(duration = 0.5, framerate = 10000)
sawtooth.make_spectrum().plot()
sawtooth.make_audio()
```

사각파와 톱니파 비교

```python
square = thinkdsp.SquareSignal().make_wave(duration=0.01, framerate=10000)
square.plot()
```

```python
square = thinkdsp.SquareSignal().make_wave(duration=0.5, framerate=10000)
square.make_spectrum().plot(high=5000)
square.make_audio()
```

사각파는 홀수배의 frequency에서 harmonics를 보여준다
또한, 사각파와 비교했을 때 톱니파는 빨리 떨어지지 않는다.

삼각파와 톱니파 비교

```python
triangle = thinkdsp.TriangleSignal().make_wave(duration=0.01, framerate=10000)
triangle.plot()                                               
```

```python
triangle = thinkdsp.TriangleSignal().make_wave(duration=0.5, framerate=10000)
triangle.make_spectrum().plot(high=5000)
triangle.make_audio()
```

삼각파는 짝수배의 frequency에서 harmonics를 보여준다
또한, 삼각파와 비교했을 때 톱니파는 빨리 떨어지지 않는다.

---

결론:
톱니파는 홀수배, 짝수배에서 모두 나타난다.
톱니파는 삼각파와 사각파와 비교했을때 비교적 완만하게 떨어진다.
