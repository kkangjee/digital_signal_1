```python
from __future__ import print_function, division

%matplotlib inline

import thinkdsp
import thinkplot

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
from IPython.display import display


```

```python
#Exercise 1.2 Go to and download a sound sample
wave = thinkdsp.read_wave('28042__bcjordan__voicedownbew.wav')
wave.make_audio()
```

```python
wave.plot()
```

```python
#Select a roughly half-second segment where the pitch is constant.
start = 0.0
duration = 0.5
segment = wave.segment(start, duration)
segment.plot()
```

low pass

```python
#Compute and plot the spectrum of the segment you selected.
spectrum = segment.make_spectrum()
spectrum.plot()
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
#Use high_pass,low_pass,band_pass and to filter out some of the harmonics.
spectrum.low_pass(880)
spectrum.plot(high=2000)
thinkplot.config(xlabel='Frequency (Hz)')

```

```python
#Then convert the spectrum back to a wave and listen to it.
filtered = spectrum.make_wave()
filtered.normalize()
filtered.apodize()
```

```python
#original sound
segment.normalize()
segment.apodize()
segment.plot()
segment.make_audio()
```

```python
#How does the sound relate to the changes you made in the spectrum?
filtered.make_audio()
```

high pass


```python
#Use high_pass,low_pass,band_pass and to filter out some of the harmonics.
spectrum.high_pass(880)
spectrum.plot(high=2000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
#Then convert the spectrum back to a wave and listen to it.
filtered = spectrum.make_wave()
filtered.normalize()
filtered.apodize()
```

```python
#How does the sound relate to the changes you made in the spectrum?
filtered.make_audio()
```

bend pass

```python
#Use high_pass,low_pass,band_pass and to filter out some of the harmonics.
spectrum.band_stop(low_cutoff=440,high_cutoff=880)
spectrum.plot(high=2000)
thinkplot.config(xlabel='Frequency (Hz)')
```

```python
#Then convert the spectrum back to a wave and listen to it.
filtered = spectrum.make_wave()
filtered.normalize()
filtered.apodize()
```

```python
#How does the sound relate to the changes you made in the spectrum?
filtered.make_audio()
```
