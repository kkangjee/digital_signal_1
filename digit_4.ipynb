```python

from __future__ import print_function, division

%matplotlib inline
import warnings
warnings.filterwarnings('ignore')

import thinkdsp
import thinkplot
import thinkstats2 

import numpy as np

from ipywidgets import interact, interactive, fixed
import ipywidgets as widgets
```

compute the spectrum of each signal.

```python
wave1 = thinkdsp.read_wave('assets/13793__soarer__north-sea.wav')
```

```python
wave2 = thinkdsp.read_wave('180929__docquesting__crowd-noise.wav')
```

```python
segment1 = wave1.segment(start=0, duration=0.5)
segment2 = wave2.segment(start=0, duration=0.5)
```

```python
spectrum1 = segment1.make_spectrum()
spectrum1.plot_power()
thinkplot.config(xlabel='Frequency (Hz)',xscale='log',yscale='log')
```

```python
spectrum2 = segment2.make_spectrum()
spectrum2.plot_power()
thinkplot.config(xlabel='Frequency (Hz)',xscale='log',yscale='log')
```

Make white noise, pink noise, and Brownian noise

```python
si_brown = thinkdsp.BrownianNoise()
w_Brown = si_brown.make_wave(duration=0.5, framerate=11025)
s_Brown = w_Brown.make_spectrum()
```

```python
si_pink = thinkdsp.PinkNoise()
w_pink = si_pink.make_wave(duration=0.5, framerate=11025)
s_pink = w_pink.make_spectrum()
```

```python
si_white = thinkdsp.UncorrelatedUniformNoise()
w_white = si_white.make_wave(duration=0.5, framerate=11025)
s_white = w_white.make_spectrum()
```

wave1번과 비교

```python
colors = ['#f18100', '#e5a7c9', '#00ff00', '#000000']#white noise는 초록색 표현

s_Brown.hs[0] = 0
s_Brown.plot_power(linewidth=1, color=colors[0])
s_pink.hs[0] = 0
s_pink.plot_power(linewidth=1, color=colors[1])
s_white.hs[0] = 0
s_white.plot_power(linewidth=1, color=colors[2])
spectrum1.hs[0] = 0
spectrum1.plot_power(linewidth=1, color=colors[3])

thinkplot.config(xlabel='Frequency (Hz)',xscale='log',yscale='log')
```

첫번째 소리는 Brownian noise와 비슷하다
